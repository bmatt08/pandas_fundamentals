```python
university_towns = []
with open('data/university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
            state = line
        else:
            university_towns.append((state, line))

university_towns[:5]
```

```python
#Setting columns as "State" and "Region Name"
towns_df = pd.DataFrame(university_towns,
                        columns = ['State', 'RegionName'])
towns_df.head()
```

```python
def get_citystate(item):
    if '(' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
towns_df = towns_df.map(get_citystate)

towns_df.head()
```

```python
#Creating a header
olympics_df = pd.read_csv('data/olympics.csv', header = 1)
olympics_df.head()
```

```python
#Renaming the columns
new_names = {'Unnamed: 0': 'Country',
            '? Summer': 'Summer Olympics',
            '01 !': 'Gold',
            '02 !': 'Silver',
            '03 !': 'Bronze',
            '? Winter': 'Winter Olympics',
            '01 !.1': 'Gold.1',
            '02 !.1': 'Silver.1',
            '03 !.1': 'Bronze.1',
            '? Games': '# Games',
            '01 !.2': 'Gold.2',
            '02 !.2': 'Silver.2',
            '03 !.2': 'Bronze.2'}

olympics_df.rename(columns = new_names, inplace = True)
olympics_df.head()
```
