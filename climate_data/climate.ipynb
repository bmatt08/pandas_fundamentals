```python
#Combining Data
```

```python
import pandas as pd
```

```python
#Importing DataSets
climate_temp = pd.read_csv("climate_temp.csv")
climate_precip = pd.read_csv("climate_precip.csv")
```

```python
climate_temp.head()
```

```python
climate_precip.head()
```

```python
#127,020 rows
#21 columns
climate_temp.shape
```

```python
#151,110 rows
#29 columns
climate_precip.shape
```

```python
#Inner Join
```

```python
#Selecting rows that only "STATION" field has GHCND:USC00045721
precip_one_station = climate_precip.query("STATION == 'GHCND:USC00045721'")
precip_one_station.head()
```

```python
precip_one_station.shape
```

```python
inner_merged = pd.merge(precip_one_station, climate_temp)
inner_merged.head()
```

```python
inner_merged.shape
```

```python
#Merging both entire datasets only on "STATION" & "DATE"
inner_merged_total = pd.merge(
    climate_temp, climate_precip, on = ["STATION", "DATE"]
)
inner_merged_total.head()
```

```python
inner_merged_total.shape
```

```python
#Outer Join
```

```python
outer_merged = pd.merge(
    precip_one_station, climate_temp, how = "outer", on = ["STATION", "DATE"]
)
outer_merged.head()
```

```python
outer_merged.shape
```

```python
#Left Join
```

```python
#Merging all DataFrame from the left 
#while removing rows from the DataFrame on the right that doesn't match any columns from the left DataFrame
left_merged = pd.merge(
    climate_temp, precip_one_station, how = "left", on = ["STATION", "DATE"]
)
left_merged.head()
```

```python
left_merged.shape
```

```python
left_merged_reversed = pd.merge(
    precip_one_station, climate_temp, how = "left", on = ["STATION", "DATE"]
)
left_merged_reversed.head()
```

```python
left_merged_reversed.shape
```

```python
#Right Join
```

```python
#Merging all DataFrame from the right 
#while removing rows from the DataFrame on the left that doesn't match any columns from the right DataFrame
right_merged = pd.merge(
    precip_one_station, climate_temp, how = "right", on = ["STATION", "DATE"]
)
right_merged.head()
```

```python
right_merged.shape
```

```python
#Combining Data across rows or columns
```

```python
#Concatenating both entire DataFrames
outer_joined = pd.concat([climate_precip, climate_temp])
outer_joined.head()
```

```python
outer_joined.shape
```

```python
#Dropping columns that have missing data
#Joining only columns that the original DataFrames have in common
inner_joined = pd.concat([climate_temp, climate_precip], join = "inner")
inner_joined.head()
```

```python
inner_joined.shape
```

```python
#Rows that have data for all columns in both DataFrames
inner_joined_cols = pd.concat(
    [climate_temp, climate_precip], axis = "columns", join = "inner"
)
inner_joined_cols.head()
```
